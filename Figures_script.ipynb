```python
# import the data here
mat = spio.loadmat('SC_metrics.mat', squeeze_me=True)
Deg = mat['degree_strength'] # array
Bet= mat['Bet_cent'] # array
Clus= mat['Clustering'] # array

# strcutural data labels
Lab = spio.loadmat('Region_labels.mat', squeeze_me=True)
Labels=Lab['Label']

# functional data metrics
mat = spio.loadmat('FC_metrics_Frontal.mat', squeeze_me=True)
PC_f = mat['Integ_BA'] # array
Q_f= mat['Integ_q'] # array
GE_f= mat['GE'] # array
Corr_f= mat['CORR'] # array

mat = spio.loadmat('FC_metrics_Temporal.mat', squeeze_me=True)
PC_t = mat['Integ_BA'] # array
Q_t= mat['Integ_q'] # array
GE_t= mat['GE'] # array
Corr_t= mat['CORR'] # array

mat = spio.loadmat('FC_metrics_OP.mat', squeeze_me=True)
PC_o = mat['Integ_BA'] # array
Q_o= mat['Integ_q'] # array
GE_o= mat['GE'] # array
Corr_o= mat['CORR'] # array

# phase-synchrony
mat = spio.loadmat('Phase_synchrony_metrics_F.mat', squeeze_me=True)
synchrony_f = mat['phase_synchrony_mean'] # array

mat = spio.loadmat('Phase_synchrony_metrics_T.mat', squeeze_me=True)
synchrony_t = mat['phase_synchrony_mean'] # array

mat = spio.loadmat('Phase_synchrony_metrics_OP.mat', squeeze_me=True)
synchrony_o = mat['phase_synchrony_mean'] # array
```

```python
# find the position of thalamic nodes

Line=4
Deg_f=np.mean(Deg[[41, 89]])
Deg_t=np.mean(Deg[[42, 90]])
Deg_o=np.mean(Deg[[43, 91]])

Bet_f=np.mean(Bet[[41, 89]])
Bet_t=np.mean(Bet[[42, 90]])
Bet_o=np.mean(Bet[[43, 91]])
                  
Clus_f=np.mean(Clus[[41, 89]])
Clus_t=np.mean(Clus[[42, 90]])
Clus_o=np.mean(Clus[[43, 91]])                             
                  
fig, ax = plt.subplots(ncols=3,nrows=1, figsize=(14,4))
ax[0].hist(Deg)
ax[0].axvline(Deg_f, color='k', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal
ax[0].axvline(Deg_t, color='r', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal  
ax[0].axvline(Deg_o, color='b', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal        
ax[1].hist(Bet)
ax[1].axvline(Bet_f, color='k', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal
ax[1].axvline(Bet_t, color='r', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal  
ax[1].axvline(Bet_o, color='b', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal   
ax[2].hist(Clus)
ax[2].axvline(Clus_f, color='k', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal
ax[2].axvline(Clus_t, color='r', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal  
ax[2].axvline(Clus_o, color='b', linestyle='dotted', linewidth=Line) # add frontal-occipitotemporal-temporal   

# set titles
ax[0].set_title('\n\nDegree\n',fontweight='bold', fontsize=20)
ax[1].set_title('\n\nBetweenness centrality\n',fontweight='bold', fontsize=20)
ax[2].set_title('\n\nClustering coefficient\n',fontweight='bold', fontsize=20)

# set legends
Temp=mpatches.Patch(color='red', label='temp')
Front=mpatches.Patch(color='k', label='front')
Oc_pariet=mpatches.Patch(color='b', label='occ_pariet')
plt.legend(handles=[Temp,Front,Oc_pariet])


plt.tight_layout()

#plt.savefig('/Users/RChenLab/Documents/TVB_Distribution/demo_scripts/Github/ThalCorProject/Figures/Fig1.png', facecolor='w', bbox_inches='tight', transparent=True,dpi=600)
```
